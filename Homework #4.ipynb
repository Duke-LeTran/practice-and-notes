```python
#Homework Assignment #4
#Authenicate and Test Permissions
#Chris Lucas
#Assistance from Duke LeTran

import pandas
import psycopg2
import time




def main():
    
    counter = 0
    
    while counter<= 4:
        
        # Exception handling to handle credential issues        
        try:
            
            #Connect to the MIMIC III database using psycopg2
            conn = psycopg2.connect("host='psql-mimic.ri.ucdavis.edu' dbname='mimic' user='mimic_ro' password='stone birthday floor spring'")
            
            print('Connection Successful, Attempting query...')
            #Exception handling to handle database query issues without crashing, currently unused but
            #kept for future positioning
            try:
            
                #Open the psycopg cursor to begin working within the database
                cur = conn.cursor()
            
                #Run a query on the MIMIC III Database, looking for the table grants given to the given user with the specific
                #privilege type
                cur.execute("""
                            SELECT *
                            FROM information_schema.role_table_grants
                            WHERE grantee = 'mimic_ro'
                            AND privilege_type = 'UPDATE'
                            """)
            
                #Save the information found in the database table to a variable
                data_mimic = cur.fetchall()
            
                #Organize the queried data into a Pandas Data Frame for modification/organization within Python
                dataframe = pandas.DataFrame(data_mimic)
            
                #Run a query on the MIMIC III database looking for the column names in the information table
                cur.execute("""
                            SELECT column_name
                            FROM information_schema.columns
                            WHERE table_schema = 'information_schema'
                            AND table_name = 'role_table_grants'
                            """)
            
                #Save the tuples generated in the query to the variable
                tuple_cols = cur.fetchall()
    
                #convert from tuples to lists of strings
                list_cols = [x[0] for x in tuple_cols]
    
                #Save the column names stored in the list as the column names for the dataframe
                dataframe.columns = list_cols
    
                #Close the cursor now that queries are complete and all data needed is saved in a variable
                cur.close()
    
                #Sort the information stored in the Dataframe
                dataframe_sorted = dataframe.sort_values(by=['table_schema', 'table_name']).reset_index(drop=True)
            
            
            
                print('The tables with UPDATE accessibility by the user: mimic_ro are printed below in the following format: Schema.TableName \n')
            
                #Print the information stored in the dataframe
                for i,row in dataframe_sorted.iterrows():
                
                    #Since the table names are associated with a specific schema, the schema and table name are joined and printed
                    #together
                    x = ".".join((row['table_schema'], row['table_name']))
                    print(x)
                
                counter = counter +     6
            except:
                    print('There was a problem carrying out your request.  Please ensure that your request uses the correct schema and table names')
                    pass
        
            
        except:
                #Counter makes it so only a specifc amount of reconnects are attempted
                counter = counter + 1
                
                print('There was a problem connecting with the server.  Waiting 5 seconds and trying again')
                #print('Attempt number:' counter'.  Five total attempts will be made')
                
                time.sleep(5)
                
    if counter == 5:
        
        print('Connection unsuccessful, please check connection and credentials before trying again')
        
    else:
        print('Query Complete')
                
    

main()
```
